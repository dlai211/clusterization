```python
import numpy as np
import matplotlib.pyplot as plt

# Define the number of bins, range, and bin edges
num_bins = 10
bin_min = -20
bin_max = 20
bins = np.linspace(bin_min, bin_max, num_bins + 1) 

bin_centers = (bins[:-1] + bins[1:]) / 2
data = bin_centers

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.hist(data, bins=bins, edgecolor='black', alpha=0.7, rwidth=0.8)
plt.xticks(bin_centers, labels=[f"{x:.1f}" for x in bin_centers])

plt.xlabel("Bin Center Positions")
plt.ylabel("Count")
plt.title("Histogram with 1 Count per Bin")
plt.grid(True)
plt.show()

```
