```python
import struct

# Binary file containing the data
binary_file = "event000000001-measurements.dat"

start_offset = 0  
bytes_to_read = 80

# Read the binary data
with open(binary_file, "rb") as bin_file:
    bin_file.seek(start_offset)
    raw_bytes = bin_file.read(bytes_to_read)
    print(len(raw_bytes))

    # Print raw bytes in hex
    print("Raw bytes:")
    print(" ".join(f"{byte:02x}" for byte in raw_bytes))

    # Test decoding at multiple positions
    print("\nTesting float decoding at different positions:")
    for i in range(len(raw_bytes) - 1):
        try:
            value = struct.unpack("2I", raw_bytes[i:i+8])[0] # unsigned long long
            print(f"Offset {start_offset + i}: {value}")
        except struct.error as e:
            print(f"Offset {start_offset + i}: Failed to decode ({e})")
```

 - geometry ID found in offset 24-31
 - measurement_id found in offset 32-39
