```python
df_tmp = dfc_annulus.merge(df_annulus, on='athena_id', how='left')
dfc_annulus = dfc[dfc.Barrel_ec != 0]
dfc_pos2_annulus = dfc[dfc.Barrel_ec == 2]
dfc_neg2_annulus = dfc[dfc.Barrel_ec == -2]
dfa_pos2_annulus = dfa[dfa.Barrel_ec == 2]
dfa_neg2_annulus = dfa[dfa.Barrel_ec == -2]

def rphiToBeam(df):
    # r = df_annulus_test.phiPitch / df_annulus_test.m_pitch0
    r = df.radius
    m_nStrips0 = df.m_nStrips0
    m_pitch0 = df.m_pitch_row
    m_lengthBF = df.m_lengthBF
    m_stereo = df.m_stereo

    phi_strip = df.local0
    b = -2 * m_lengthBF * np.sin(m_stereo/2 + phi_strip)
    c = m_lengthBF*m_lengthBF - r*r
    rad_strip = (-b + np.sqrt(b*b - 4*c))/2
#     rad_strip = rad_strip + df.lorentzShift

    # stripToBeam 
    m_waferCentreR = df.m_waferCentreR
    x_strip = rad_strip * np.cos(phi_strip)
    y_strip = rad_strip * np.sin(phi_strip)
    # x_strip = df_annulus_test.x_strip
    # y_strip = df_annulus_test.y_strip
    x_beam = np.cos(m_stereo) * (x_strip - m_waferCentreR) - np.sin(m_stereo) * y_strip + m_waferCentreR
    y_beam = np.sin(m_stereo) * (x_strip - m_waferCentreR) + np.cos(m_stereo) * y_strip
    return y_beam, x_beam


def diff(dfc_annulus, dfa_annulus):
    gid = np.array(list(Counter(sorted(dfc_annulus.geometry_id)).keys()))
    diff = np.array(list(Counter(sorted(dfc_annulus.geometry_id)).values())) - np.array(list(Counter(sorted(dfa_annulus.geometry_id)).values()))
    index = diff==0
    gid2 = gid[index]
    annulus_r_diff, annulus_phi_diff, annulus_x_diff, annulus_y_diff = [], [], [], []
    for i in tqdm(gid2):
        r_c = np.array(dfc_annulus[dfc_annulus['geometry_id'] == i].local1)
        phi_c = np.array(dfc_annulus[dfc_annulus['geometry_id'] == i].local0)
        shift_c = np.array(dfc_annulus[dfc_annulus['geometry_id'] == i].lorentzShift)
        angleshift_c = np.array(dfc_annulus[dfc_annulus['geometry_id'] == i].angleShift)
        x_c, y_c = rphiToBeam(df_tmp[df_tmp['geometry_id'] == i])
        
        shiftb_c = np.array(dfc_annulus[dfc_annulus['geometry_id'] == i].lorentzShift)
    
        x_a = np.array(dfa_annulus[dfa_annulus['geometry_id'] == i].local0)
        y_a = np.array(dfa_annulus[dfa_annulus['geometry_id'] == i].local1)
        r_a = cal_r(x_a, y_a)
        phi_a = np.arctan2(y_a, x_a)
        annulus_r_diff.append(np.sort(r_c) - np.sort(r_a))
        annulus_phi_diff.append(np.sort(phi_c) - np.sort(phi_a))
        annulus_x_diff.append(np.sort(x_c + shiftb_c) - np.sort(x_a))
        annulus_y_diff.append(np.sort(y_c) - np.sort(y_a))
    return np.concatenate(annulus_r_diff), np.concatenate(annulus_phi_diff), np.concatenate(annulus_x_diff), np.concatenate(annulus_y_diff)

annulus_r_pos2_diff, annulus_phi_pos2_diff, annulus_x_pos2_diff, annulus_y_pos2_diff = diff(dfc_pos2_annulus, dfa_pos2_annulus)
annulus_r_neg2_diff, annulus_phi_neg2_diff, annulus_x_neg2_diff, annulus_y_neg2_diff = diff(dfc_neg2_annulus, dfa_neg2_annulus)

```
